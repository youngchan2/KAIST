## **NumPy**

1. NumPy is the fundamental package for scientific computing in Python and is used to perform computations on multi-dimensional data easily and effectively.
2. Numpy provides a new data structure called arrays which allow efficient vector and matrix operations and a number of linear algebra operations

### Import NumPy package



````python
#Import numpy library
import numpy as np
````

### Scalar, Vector, Matrix, Tensor
```np.array``` is a central data structure of the NumPy library

````python
# define a scalar
a = np.array(1)
print('a = ', a.__repr__())
print('dimension of a: ', a.ndim)
print('shape of a: ', a.shape)

# define a vector
b = np.array([1., 2., 3.])
print('\nb = ', b.__repr__())
print('dimension of b: ', b.ndim)
print('shape of b: ', b.shape)

# define a matrix
c = np.array([[1., 2., 3.], [4., 5., 6.]])
print('\nc = ',c.__repr__())
print('dimension of c: ', c.ndim)
print('shape of c: ', c.shape)

# define a tensor (N-dimensional array)
d = np.array([[[[1., 2., 3.], [1., 2., 3.], [1., 2., 3.]],
               [[4., 5., 6.], [4., 5., 6.], [4., 5., 6.]]],
              [[[7., 8., 9.], [7., 8., 9.], [7., 8., 9.]],
               [[10., 11., 12.], [10., 11., 12.], [10., 11., 12.]]]])
print('\nd = ',d.__repr__())
print('dimension of d: ', d.ndim)
print('shape of d: ', d.shape)
````

### Shape Manipulation

````python
# reshape

a = np.arange(10)
print(a, a.shape)

b = a.reshape(2,5)
print(b, b.shape)
````

````python
# add an additional dimension
print(a.shape)
print(a[None].shape)  # add extra dimension in first dim
print(b[None].shape)
````

````python
# add in second dim
print(a[:, None].shape)
print(b[:, None].shape)
````

````python
# add in last dim
print(a[..., None].shape)
print(b[..., None].shape)
````

````python
# add in second last dim
a[..., None, :].shape
````

````python
# stack
a = np.ones((3,2))
b = np.zeros((3,2))
print(np.vstack([a, b]))  #a 뒤에 b가 붙음
print(np.hstack([a, b]))  #a 행 한개 뒤에 b 행 한개 붙음 / a와 b 행 개수 다르면 오류
````

````python
# concatenate
print(np.concatenate([a, b], axis=0))
print(np.concatenate([a, b], axis=1)) #axis는 dimension 수까지 있음
````

````python
# matrix transpose
a = np.arange(10).reshape(2,5)
print(a)
print(a.T)
print(a.transpose())
````

````python
# tensor transpose
a = np.arange(24).reshape(2,3,4)
print(a, a.shape)

# swap axis 0 and 2
# axis number에 해당 axis = 0 -> 2 / axis = 1 ->3 / axis = 2 ->4
# 2,1,0은 (4,3,2) 행렬 의미
b = np.transpose(a, [2, 1, 0])
print(b, b.shape)
````

### Numpy Operations (Math)

````python
# Basic mathematical functions in the numpy module are available and operate elementwise on arrays.
# support all basic numerical operations such as +. -. *, /, ** ..
a = np.arange(0, 3, 1)
b = np.arange(1, 4, 1)
print('a = ', a.__repr__())
print('b = ', b.__repr__(), end='\n')

# 각 원소에 연산 적용
print('a = ', a)
print('a + 5 = ', a + 5)
print('a^2 = ', a ** 2)
print('sin(a) = ', np.sin(a))
print('logical operation of a < 1: ', a < 1)
````

````python
# operator * is not vector multiplication but elementwise multiplication.
print('array a = ', a)
print('array b = ', b)
a * b
````

````python
# for vector multiplication, we use the dot function to compute inner products of vectors
# 0*1 + 1*2 + 2*3 = 8
np.dot(a, b), a@b
````

````python
# Matrix multiplication

a = np.array([[1,2],[3,4]])
b = np.array([[4,3],[2,1]])

np.matmul(a,b), np.dot(a,b), a@b
````

````python
# using np.ndarray.method()
a = np.arange(6).reshape(2,3)
print('array a = ', a)
print('max value of each column: ', a.max(axis = 0)) # max of each column
print('min value of each row:', a.min(axis = 1)) # min of each row
print('sums of all elements:', a.sum()) # sum of all elements
print('sums of each row:', a.sum(axis = 1)) # sum of each row
print('max value of array (matrix) c:', a.max()) # max of c
````

````python
# take operation while keeping dimension
np.sum(a, axis=0, keepdims=True)
````

````python
np.sum(a, axis=1, keepdims=True)
````

### Broadcasting
Broadcasting is a powerful mechanism that allows numpy to work with arrays of different shapes when computing mathematical operations.

````python
# Vector and scalar
a = np.arange(3)
b = 2.

print("a = ", a, "/ b = ", b)
print("a+b = ", a+b)
print("a-b = ", a-b)
print("a*b = ", a*b)
print("a/b = ", a/b)
````

````python
# Matrix and vector
a = np.arange(1, 7).reshape(2, 3)
b = np.arange(1, 4)

print("a = ", a, "/ b = ", b)
print("a+b = ", a+b)
print("a-b = ", a-b)
print("a*b = ", a*b)
print("a/b = ", a/b)
````

````python
# Tensor and matrix
a = np.arange(1, 13).reshape(2, 2, 3)
b = np.arange(1, 7).reshape(2, 3)

print("a = ", a, "/ b = ", b)
print("a+b = ", a+b)
print("a-b = ", a-b)
print("a*b = ", a*b)
print("a/b = ", a/b)
````

### Copy in numpy

3 cases of copying NumPy array

````python
# Case 1
# a와 b 같은 array를 pointing
# a나 b 둘 중 하나로 array를 update 하면 둘 다 update 됨

a = np.zeros((2,2))
b = a #No copy at all # Share both the data and properties(e.g., dimension of array)
print('b: \n', b.__repr__())

b[1,1] = 1
print('b: \n', b.__repr__())
print('a: \n', a.__repr__()) # a is also changed


b.shape = (1,4)
print('shape of a: ', a.shape) #The shape of a is also changed
print('shaep of b: ', b.shape)
````

````python
# Case 2 : Shallow copy
# data만 바뀌고 차원은 각자 유지
a = np.zeros((2,2))
b = a.view()                    #Shallow copy # Share the data but not properties(e.g., dimension of array)
print('b: \n', b.__repr__())

b[1,1] = 1
print('b: \n', b.__repr__())
print('a: \n', a.__repr__())    # a is also changed!!

b.shape = (1,4)
print('shape of a: ', a.shape)  #The shape of a is not changed
print('shaep of b: ', b.shape)
````

````python
# Case 3 : Deep copy
# a와 관련없이 a의 모든걸 복사해감

a=np.zeros((2,2))
c = a.copy()                  #Deep copy # Create an independet variable not sharing both the data and properties
print('c: \n', c.__repr__())

c[1,1] =1
print('c: \n', c.__repr__())
print('a: \n', a.__repr__())   # a is not changed
````

## Plotting data

Matplotlib is a popular libarary for creating static, animated, interactive visualizations in Python.

### Import matplotlib

````python
import matplotlib.pyplot as plt
````

### Generate data & Define a function

````python
x_temp = np.arange(0, 10, 0.01) # (start, end, size) size: 간격
x = np.linspace(0, 10, 1000)  # (start, end, num) num: 개수 / 개수에 맞게 간격 조정됨
print(x_temp.shape, x.shape)

y = 3*x**2 - 20*x + 25
````

### Plot the function

````python
plt.plot(x, y, label='3x^2 - 20x + 25', color='red')
plt.grid()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend() #범례
plt.show()
````

````python
plt.scatter(x, y, label='3x^2 - 20x + 25', color='green', s=1) #s는 점의 size
plt.grid()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
````

We are going to generate 100 data points from each of 3 multivariate normal distributions:
\begin{align*}
&X\sim\mathcal{N}(\mu, \Sigma),\\
&\text{where}\quad \mu_1=\begin{bmatrix} 0 & 10 \end{bmatrix}, \quad \mu_2=\begin{bmatrix} 10 & 0 \end{bmatrix},\quad \mu_3=\begin{bmatrix} 20 & 20 \end{bmatrix}, \quad\Sigma=\begin{bmatrix} 3 & 0 \\ 0 & 3 \end{bmatrix}.
\end{align*}

Now plot the data and the centroid of data sampled from each of the distributions.

### **Exercise 1**. generate 1000 data points from each normal distribution (use ```np.random.normal```)

````python
n = 1000
###########################

# Enter your code
mu1 = np.array([0, 10])
mu2 = np.array([10, 0])
mu3 = np.array([20, 20])
mu = np.array([mu1, mu2, mu3])

x11 = np.random.normal(mu[0][0], 3**0.5, n)
x12 = np.random.normal(mu[0][1], 3**0.5, n)
x1 = np.array([x11, x12])

x21 = np.random.normal(mu[1][0], 3**0.5, n)
x22 = np.random.normal(mu[1][1], 3**0.5, n)
x2 = np.array([x21, x22])

x31 = np.random.normal(mu[2][0], 3**0.5, n)
x32 = np.random.normal(mu[2][1], 3**0.5, n)
x3 = np.array([x31, x32])

###########################
print(x1.shape)
print(x2.shape)
print(x3.shape)
````

### **Exercise 2**. compute the mean and covariance for x1, x2, x3

````python
# Enter your code
mu1 = np.array([np.mean(x11), np.mean(x12)])
mu2 = np.array([np.mean(x21), np.mean(x22)])
mu3 = np.array([np.mean(x31), np.mean(x32)])
print("mean1: ", mu1)
print("mean2: ", mu2)
print("mean3 ", mu3)
print('\n')
#var11 = x11 - mu1[0]
#var12 = x12 - mu1[1]
#var1 = np.array([var11, var12])
#cov1 = var1@var1.T/10000
cov1 = np.cov([x11, x12])

#var21 = x21 - mu2[0]
#var22 = x22 - mu2[1]
#var2 = np.array([var21, var22])
#cov2 = var2@var2.T/10000
cov2 = np.cov([x21, x22])

#var31 = x31 - mu3[0]
#var32 = x32 - mu3[1]
#var3 = np.array([var31, var32])
#cov3 = var3@var3.T/10000
cov3 = np.cov([x31, x32])
print("cov1: ", cov1)
print("cov2: ", cov2)
print("cov3: ", cov3)
````
